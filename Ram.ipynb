```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generating synthetic data
X, _ = make_blobs(n_samples=300, centers=8, cluster_std=0.60, random_state=56)

# Creating the model
kmeans = KMeans(n_clusters=8)

# Fitting the model
kmeans.fit(X)

# Getting the cluster centers and labels
centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Plotting the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='g', s=200, alpha=0.75)
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Sample Data Loading
# Replace this with actual healthcare data
data = {
    'age': [25, 34, 28, 52, 45, 36, 29, 56, 42, 50],
    'bmi': [22.4, 27.1, 26.3, 30.0, 29.5, 28.1, 27.8, 31.2, 26.5, 29.9],
    'blood_pressure': [120, 135, 128, 142, 139, 132, 130, 145, 138, 140],
    'cholesterol': [190, 210, 200, 230, 225, 215, 208, 240, 220, 230],
    'readmission': [0, 1, 0, 1, 1, 0, 0, 1, 0, 1]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Features and Target
X = df[['age', 'bmi', 'blood_pressure', 'cholesterol']]
y = df['readmission']

# Splitting the Data into Training and Testing Sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

### Logistic Regression
# Model Initialization
log_model = LogisticRegression()

# Model Training
log_model.fit(X_train, y_train)

# Predictions
log_y_pred = log_model.predict(X_test)

# Evaluation
print("Logistic Regression Evaluation:")
print("Accuracy:", accuracy_score(y_test, log_y_pred))
print("\nClassification Report:\n", classification_report(y_test, log_y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, log_y_pred))

### Linear Regression
# Model Initialization
lin_model = LinearRegression()

# Model Training
lin_model.fit(X_train, y_train)

# Predictions
lin_y_pred = lin_model.predict(X_test)

# Convert continuous predictions to binary using a threshold (e.g., 0.5)
lin_y_pred_binary = [1 if pred > 0.5 else 0 for pred in lin_y_pred]

# Evaluation
print("\nLinear Regression Evaluation (after thresholding):")
print("Accuracy:", accuracy_score(y_test, lin_y_pred_binary))
print("\nClassification Report:\n", classification_report(y_test, lin_y_pred_binary))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, lin_y_pred_binary))
```
